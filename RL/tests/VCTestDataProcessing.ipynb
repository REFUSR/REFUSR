## Revising Data from Lucca 

```python
import pandas as pd

COLNAMES = ['num_data', 'num_registers', 'output_registers', 'program', 'num_cases', 'input_data', 'output']

df = pd.read_csv('data/data.csv', index_col=False, names=COLNAMES)
```

```python
def access_bit(data, num):
    base = int(num // 8)
    shift = int(num % 8)
    return (data[base] >> shift) & 0x1
```

```python
import struct
from io import BytesIO
import numpy as np

input_dataxs = []
output_dataxs = []
for i in range(df.shape[0]):  
    x = bytearray(struct.pack("f", df['input_data'].iloc[i]))  
    y = bytearray(struct.pack("f", df['output'].iloc[i]))  
    

    #some detective work, it seems like the same input- all of the register state
    bx = [access_bit(x,i) for i in range(len(x)*8)]
    by = [access_bit(y,i) for i in range(len(y)*8)]
    
    bx = " ".join(map(str, bx))
    by = " ".join(map(str, by))
    
    assert len(bx) == len(by)

    
    input_dataxs.append(bx)
    output_dataxs.append(by)
```

```python
df['input_data'] = input_dataxs 
```
